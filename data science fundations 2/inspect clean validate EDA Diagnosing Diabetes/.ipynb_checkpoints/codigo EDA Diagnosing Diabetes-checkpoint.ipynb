```python
import pandas as pd
import numpy as np

diabetes_data = pd.read_csv('diabetes.csv')
print(diabetes_data.head())
print(diabetes_data.dtypes)#outcome es object y no booleano como debería
print(len(diabetes_data))#768 registros
print(diabetes_data.info())#no hay nulos, igual hay que ver el outcome que es string 

print(diabetes_data.describe())#raro que sean muchos de 33 años con un promedio de 4 embarazos
#no podes tener 0 de glucosa, presion o de los otros datos
#es mucho 17 embarazos maximos
#es mucho 846 de insulina

##HASTA ACÁ NOS DIMOS CUENTA QUE LOS DATOS NO ESTÁN LIMPIOS A PESAR QUE NO HAYAN NULOS

#pasamos los ceros a nulos, como corresponde
diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
#print(diabetes_data.info())

diabetes_null = diabetes_data[ diabetes_data.isnull().any(axis=1)]
print("Diabetes null:")
print(diabetes_null)
#print(diabetes_nulls.head(30)) #varios tienen mas de un dato nulo, tal vez algo pasó al momento de hacer mediciones
#si un dato falla, los de la derecha también (no el BMI)

#vamos a ver que pasa cojn outcome
print(diabetes_data['Outcome'].unique())
diabetes_data['Outcome']= diabetes_data['Outcome'].replace('O', '0')
diabetes_data['Outcome']= diabetes_data['Outcome'].astype("int")
print(diabetes_data['Outcome'].unique())
```
