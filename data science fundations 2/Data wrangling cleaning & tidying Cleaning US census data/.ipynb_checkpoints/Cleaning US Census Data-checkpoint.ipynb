1. The first visualization your boss wants you to make is a scatterplot that shows average income in a state vs proportion of women in that state.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

files = glob.glob('states*.csv')


2. It will be easier to inspect this data once we have it in a DataFrame. You can't even call `.head()` on these `csv`s! How are you supposed to read them?

   Using `glob`, loop through the census files available and load them into DataFrames. Then, concatenate all of those DataFrames together into one DataFrame, called something like `us_census`.

````python
#no podemos manipular la info como data frames, así que haremos un arreglo de data frames
dfs = []
for state in files:
    data = pd.read_csv(state)
    dfs.append(data)#los podemos unir porque tienen las mismas columnas los diferentes archivos
us_census = pd.concat(dfs)#unimos las dfs en una sola
us_census.reset_index()
````

## Regex to the Rescue

5. Use regex to turn the `Income` column into a format that is ready for conversion into a numerical type.

````python
us_census.Income = us_census['Income'].str.replace('\$','',regex=True)


    
print(us_census.head())
````

6. Look at the `GenderPop` column. We are going to want to separate this into two columns, the `Men` column, and the `Women` column.

   Split the column into those two new columns using `str.split` and separating out those results.

````python
string_split = us_census['GenderPop'].str.split('_', expand=True)
us_census['Men'] = string_split[0]
us_census['Women'] = string_split[1]
us_census = us_census.drop(columns=['GenderPop'])
print(us_census.head())
````

````python
print(us_census.Women)
````

7. Convert both of the columns into numerical datatypes.

   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

````python
#i'll also correct the other columns values
us_census_new = us_census#good practise to make a new df when whe change the data

for column in us_census.columns[3:-2]: # since Hispanic
    us_census_new[column] = us_census[column].str.replace('($|%| |,)', '', regex=True)#take out rare character such as '$' and '%' and ' '
    us_census_new[column] = us_census_new[column].astype('float')
    
us_census_new['Men'] = us_census_new.Men.str.replace('M','',regex=False)
us_census_new['Men'] = us_census_new['Men'].astype('int64') 

us_census_new['Women'] = us_census_new.Women.str.replace('F','',regex=False)  
us_census_new['Women'] = us_census_new['Women'].replace('', '0')#sometimes there's no info for girls, later we will fix this
us_census_new['Women'] = us_census_new['Women'].astype('int64') 

print(us_census_new)
print(us_census_new.dtypes)
````

8. Now you should have the columns you need to make the graph and make sure your boss does not slam a ruler angrily on your desk because you've wasted your whole day cleaning your data with no results to show!


````python
plt.scatter(us_census_new.Women, us_census_new.Income)
plt.show()
plt.clf()
````

9. You want to double check your work. You know from experience that these monstrous csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

   We can fill in those `nan`s by using pandas' `.fillna()` function.
   
   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
print(us_census_new.info())#pacific tiene 5 nulos
print(us_census_new.head(20))
us_census_new.Pacific = us_census_new.Pacific.fillna(100 - us_census_new.Hispanic - us_census_new.White - us_census_new.Black - us_census_new.Native - us_census_new.Asian)
print(us_census_new.info())#pacific tiene 5 nulos
print(us_census_new.head(20))
````

````python
print(us_census_new.Women[us_census_new.Women == 0])
us_census_new.Women = us_census_new.apply(lambda row: row.TotalPop - row.Men if row.Women == 0 else row.Women, axis=1)#we cannot use replace because that function is for strings
print(us_census_new.Women[us_census_new.Women == 0])
````

14. Try to make a histogram for each one!

    You will have to get the columns into the numerical format, and those percentage signs will have to go.
    
    Don't forget to fill the `nan` values with something that makes sense! You probably dropped the duplicate rows when making your last graph, but it couldn't hurt to check for duplicates again.

````python
#histogram compares quantitative data. We're gonna make one hist for each race on every rows
#i've already completed the nan values por the Pacific column

for race in us_census_new.columns[2:8]:
    plt.hist(us_census_new[race], range=(0,100), bins=len(us_census_new))
    plt.xlabel('Percentage') 
    plt.ylabel('Frequency')
    plt.title(race+" percentage on different states")
    plt.xticks(list(range(0, 101, 10)))
    plt.yticks(list(range(0,len(us_census_new)+1,5)))
    plt.show()
    plt.clf()
````
